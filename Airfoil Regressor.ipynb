# Now we will do Feature Engineering

```python
df1 = df.copy()
```

```python
outliers = ['frequency', 'suction_side', 'pressure_level']
```

```python
outliers_df = df1[outliers]
```

```python
sk = outliers_df.skew(), outliers_df.kurtosis()
```

```python
sk
```

#### 1 - Since skewness and kurtosis is not so large, so there is no need of any transformation.
#### 2 - Outliers doesn't create much affect in our dataset
#### 3 - Now we will check whether feature scaling is required or not ?
#### 4 - We will also train our model in this duration.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
std = StandardScaler()
```

```python
X = df1.iloc[ : , :-1]
y = df1.iloc[: ,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
train_scaler = std.fit(X_train)
```

```python
train_scaler.mean_
```

```python
X_train_scaled = std.transform(X_train)
X_test_scaled = std.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
#### Before Scaling

sns.kdeplot(X_train['frequency'])
sns.kdeplot(X_train['chord_length'])
```

```python
#### After Scaling

sns.kdeplot(X_train_scaled['frequency'])
sns.kdeplot(X_train_scaled['chord_length'])
```

# Training the model - Linear, Ridge, Lasso Regression 

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr_scaled = LinearRegression()
```

```python
lr.fit(X_train, y_train)
lr_scaled.fit(X_train_scaled, y_train)
```

```python
y_predict = lr.predict(X_test)
y_predict_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
print('Actual', mean_squared_error(y_test, y_predict), mean_absolute_error(y_test, y_predict), r2_score(y_test, y_predict))
print('Scaled', mean_squared_error(y_test, y_predict_scaled), mean_absolute_error(y_test, y_predict_scaled), r2_score(y_test, y_predict_scaled))
```

### From above observations we can find out that before and after feature scaling our model gives same result. So it is not necessary to do feature scaling.


```python
from sklearn.linear_model import Ridge, Lasso
```

```python
rd = Ridge()
ls = Lasso()
```

```python
rd.fit(X_train, y_train)
ls.fit(X_train, y_train)
```

```python
predict_ridge = rd.predict(X_test)
predict_lasso = ls.predict(X_test)
```

```python
print('Ridge', mean_squared_error(y_test, predict_ridge), mean_absolute_error(y_test, predict_ridge), r2_score(y_test, predict_ridge))
print('Lasso', mean_squared_error(y_test, predict_lasso), mean_absolute_error(y_test, predict_lasso), r2_score(y_test,predict_lasso))
```
